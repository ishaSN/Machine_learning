### Outlier removal

```python
data.dtypes
```

```python
dict = {}
for col in ['DurationOfPitch','MonthlyIncome']:
    percentile25 = data[col].quantile(0.25)
    percentile75 = data[col].quantile(0.75)
    IQR = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * IQR
    lower_limit = percentile25 - 1.5 * IQR
    dict['upper_limit' + '_' + col] = upper_limit
    dict['lower_limit' + '_' + col] = lower_limit
```

```python
dict
```

```python
for col in ['DurationOfPitch','MonthlyIncome']:
    data[col] = np.where(
        data[col] > dict['upper_limit_' + col],
        dict['upper_limit_' + col],
        np.where(
           data[col] < dict['lower_limit_' + col],
           dict['lower_limit_' + col],
           data[col]
    )
)   
```

After removing the ourliers, the distribution plot is spread more.

```python
plt.figure(1)
plt.subplot(121)
#sns.distplot(data['DurationOfPitch'])
sns.histplot(data["DurationOfPitch"], kde=True);
plt.title('Distribution of DurationOfPitch');
plt.subplot(122)
data['DurationOfPitch'].plot.box(figsize=(16,5))
plt.show()
```
